```python
import pandas as pd
import quilt3 as q3
import re
```

```python
#grep "%" 'hurdat2-2000-2023-051124_2konwards.txt'
df = pd.read_csv(
   'hurdat2-2000-2023-051124_2konwards.txt',
    sep="%",
    header=None
)
```

```python
widths = set()
for s in df[0]:
    widths.add(len(s.split(',')))
```

```python
storms = {}
header = None
for s in df[0]:
    tokens = s.split(',')
    assert (len(tokens) == 4) or len(tokens) == 21
    if len(tokens) == 4:
        header = s
        assert not (header in storms)
        storms[header] = []
    else:
        assert header
        row = [t.strip() for t in tokens]
        storms[header].append(row)
```

```python
frames = []
for k in storms:
    code, name, entries, _ = [t.strip() for t in k.split(',')]
    columns = {
        'basin': code[:2],
        'ATCF Cyclone Number': code[2:4],
        'year': code[4:8],
        'name': name,
        'num track entries': entries
        
    }
    table = pd.DataFrame(storms[k], columns = [
    'date', 'time', 'record', 'storm_type', 'latitude', 'longitude', 
    'max_sustained_wind(knots)', 'min_pressure(mbar)',
    '34kt_wind_radii_NE', '34kt_wind_radii_SE', '34kt_wind_radii_SW', '34kt_wind_radii_NW', 
    '50kt_wind_radii_NE', '50kt_wind_radii_SE', '50kt_wind_radii_SW', '50kt_wind_radii_NW', 
    '64kt_wind_radii_NE', '64kt_wind_radii_SE', '64kt_wind_radii_SW', '64kt_wind_radii_NW', 
    'radii_max_wind(nautical_miles)'
])
    # add header columns
    for c in columns:
        table[c] = columns[c]
    table
    
    frames.append(table)
# concat arrays of arrays in ONE shot, making ONE dataframe = muuuch faster
df = pd.concat(frames)
df.head(10)
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df['time'] = pd.to_datetime(df['time'], format='%H%M').dt.time
df['year'] = pd.to_datetime(df['year'], format='%Y')

```

```python
#Other colkumns to convert to float
columns_to_convert = [
    'max_sustained_wind(knots)',
    'min_pressure(mbar)',
    '34kt_wind_radii_NE',
    '34kt_wind_radii_SE',
    '34kt_wind_radii_SW',
    '34kt_wind_radii_NW',
    '50kt_wind_radii_NE',
    '50kt_wind_radii_SE',
    '50kt_wind_radii_SW',
    '50kt_wind_radii_NW',
    '64kt_wind_radii_NE',
    '64kt_wind_radii_SE',
    '64kt_wind_radii_SW',
    '64kt_wind_radii_NW',
    'radii_max_wind(nautical_miles)'
]

df[columns_to_convert] = df[columns_to_convert].astype(float)
```
